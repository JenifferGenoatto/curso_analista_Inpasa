# Criar colunas no arquivo

```python
df['Ano'] = df['Arrival_Date'].apply(lambda x : x.year)
df['Mês'] = df['Arrival_Date'].apply(lambda x : x.month)
df['Dia'] = df['Arrival_Date'].apply(lambda x : x.day)
df['Dia da Semana'] = df['Arrival_Date'].apply(lambda x : x.weekday())
```

```python
df.head()
```

```python
#pd.concat([df['Market'], df.drop('Market', axis=1)], axis=1) #inverter colunas
```

```python
def dia_da_semana(dia):
    if dia == 0:
        return 'Segunda-feira'
    if dia == 1:
        return 'Terça-feira'
    if dia == 2:
        return 'Quarta-feira'
    if dia == 3:
        return 'Quinta-feira'
    if dia == 4:
        return 'Sexta-feira'
    if dia == 5:
        return 'Sábado'
    if dia == 6:
        return 'Domingo'
```

```python
df['Dia da Semana'] = df['Dia da Semana'].apply(dia_da_semana)  # troca os numeros pelo dia da semana escrito
```

```python
df.head()
```

```python
df['Diferença Preço (max-min)'] = df['Max Price'] - df['Min Price']
```

```python
df.head()
```

```python
df['Modal Price 15%'] = df['Modal Price'] * 0.15
```
